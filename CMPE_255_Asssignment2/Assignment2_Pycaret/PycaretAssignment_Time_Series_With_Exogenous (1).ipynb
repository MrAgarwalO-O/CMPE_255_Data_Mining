```python
import time
import numpy as np
import pandas as pd

url = '/content/simulated_sales_data.csv'
data = pd.read_csv(url)
data
```

```python
from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment

fh = 12
fold = 3

fig_kwargs = {
    # "renderer": "notebook",
    "renderer": "png",
    "width": 1000,
    "height": 600,
}
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
dt_target = 'Sales'
```

```python
eda = TSForecastingExperiment()
eda.setup(data, target=dt_target, fh=fh, fig_kwargs=fig_kwargs, use_gpu=True)
```

```python
eda.check_stats()
```

```python
best_model = eda.compare_models()
```

```python
aarima_model = eda.create_model('auto_arima')
aarima_model_finalized = eda.finalize_model(aarima_model)
```

```python
holdout_pred = eda.predict_model(best_model)
```
