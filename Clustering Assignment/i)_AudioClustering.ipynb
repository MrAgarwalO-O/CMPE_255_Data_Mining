**Setup**

```python
import os
import zipfile
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
```

**Loading the audio files**

```python
with zipfile.ZipFile('/content/set_a.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/data/dataset')

url = '/content/data/dataset'
```

```python
def load_audio_file(url):
    audio, sample_rate = librosa.load(url, sr=None)
    return audio, sample_rate
```

**Function to plot waveform and spectrogram**

```python
def plot_audio_analysis(audio, sample_rate):
    # Plot waveform
    plt.figure(figsize=(12, 4))
    librosa.display.waveshow(audio, sr=sample_rate)
    plt.title('Waveform')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()

    # Plot spectrogram
    plt.figure(figsize=(12, 4))
    spectrogram = librosa.stft(audio)
    spectrogram_db = librosa.amplitude_to_db(np.abs(spectrogram))
    librosa.display.specshow(spectrogram_db, sr=sample_rate, x_axis='time', y_axis='hz')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.show()
```

**Loading and displaying a sample audio file**

```python
sample_file = os.path.join(url, os.listdir(url)[2])
audio, sample_rate = load_audio_file(sample_file)
```

**Playing the audio**

```python
ipd.Audio(audio, rate=sample_rate)
```

**Analyze the audio**

```python
plot_audio_analysis(audio, sample_rate)
```
